```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
import warnings
warnings.filterwarnings(action='ignore')

# Read the data
def load_data():
    X_full = pd.read_csv('housing_dat.csv', index_col='Id')
    return X_full

X_full = load_data()
```

```python
# Separating target column from rest of data
y = X_full.pop('SalePrice')
y.head()
```

```python
# Break off validation set from training data
X_train_full, X_valid_full, y_train, y_valid = train_test_split(X_full, y, 
                                                                train_size=0.8, test_size=0.2,
                                                                random_state=0)

# Selecting categorical columns with relatively low cardinality
categorical_cols = [colname for colname in X_train_full.columns if
                    X_train_full[colname].nunique() < 10 and 
                    X_train_full[colname].dtype == "object"]

# Select numerical columns
numerical_cols = [colname for colname in X_train_full.columns if 
                X_train_full[colname].dtype in ['int64', 'float64']]

# All chosen columns
selected_cols = categorical_cols + numerical_cols
X_train = X_train_full[selected_cols].copy()
X_valid = X_valid_full[selected_cols].copy()
```

```python
# Data Cleaning 
# Importing necessary tools

from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder
from sklearn.pipeline import FeatureUnion
from sklearn.base import BaseEstimator, TransformerMixin
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.compose import ColumnTransformer

s = ['mean', 'median', 'most_frequent', 'constant']
```

```python
TotRmsAbvGrd_index = X_train[numerical_cols].columns.get_loc("TotRmsAbvGrd")
FirstFlrSF_index = X_train[numerical_cols].columns.get_loc("1stFlrSF")
SecondFlrSF_index = X_train[numerical_cols].columns.get_loc("2ndFlrSF")

# Creating a class to add an extra feature
class addFeature(BaseEstimator, TransformerMixin):
    def __init__ (self, add_rooms_per_sqft_abvgrd = True):
        self.add_rooms_per_sqft_abvgrd = add_rooms_per_sqft_abvgrd;
    def fit(self, X, y=None):
        return self
    def transform(self, X, y=None):
        if self.add_rooms_per_sqft_abvgrd:
            rooms_per_sqft_abvgrd = X[:, TotRmsAbvGrd_index] / (X[:, FirstFlrSF_index]+X[:, SecondFlrSF_index])
            return np.c_[X, rooms_per_sqft_abvgrd]
        

        
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import cross_val_score 

# From research, RandomeForestRegressors work well for this data
# Here we try different imputer strategies and find the best one using cross validation 
# We select the strategy that gives the lowest root mean squared error
strategies = ['mean', 'median', 'most_frequent', 'constant']
rmse_scores = []
def display_scores(scores):
    print("Scores: ", scores)
    print("Mean: ", scores.mean())
    print("SD: ", scores.std())
    
for s in strategies:
    # Pipeline to preprocess numerical data
    numerical_pipeline = Pipeline([
        ('imputer', SimpleImputer(strategy = s)),
        ('feature_adder', addFeature()),
        ('scaler', StandardScaler()),
    ])
    # Pipeline to preprocess categorical data
    categorical_pipeline = Pipeline(steps=[
        ('imputer', SimpleImputer(strategy='most_frequent')),
        ('onehot', OneHotEncoder(handle_unknown='ignore'))
    ])
    # Bundle preprocessing for numerical and categorical data
    preprocessor = ColumnTransformer(
        transformers=[
            ('num', numerical_pipeline, numerical_cols),
            ('cat', categorical_pipeline, categorical_cols)
        ])
    pipeline = Pipeline(steps=[('p', preprocessor), ('m', RandomForestRegressor(random_state=0))]) 
    scores = cross_val_score(pipeline, X_train, y_train, scoring='neg_mean_squared_error', cv=10)
    rmse_scores.append(np.sqrt(-scores))
```

```python
for strat, scores in zip(strategies, rmse_scores):
    print(strat)
    display_scores(scores)
```

```python
# From the results in the cell above it looks like the constant strategy 
# gives the best results 
# Pipeline to preprocess numerical data
numerical_pipeline = Pipeline([
    ('imputer', SimpleImputer(strategy = 'constant')),
    ('feature_adder', addFeature()),
    ('scaler', StandardScaler()),
])
# Pipeline to preprocess categorical data
categorical_pipeline = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='most_frequent')),
    ('onehot', OneHotEncoder(handle_unknown='ignore'))
])
# Bundle preprocessing for numerical and categorical data
preprocessor = ColumnTransformer(
    transformers=[
        ('num', numerical_pipeline, numerical_cols),
        ('cat', categorical_pipeline, categorical_cols)
    ])

X_prep = preprocessor.fit_transform(X_train)
```

```python
from sklearn.model_selection import GridSearchCV

# Let's try to use grid search to find the best hyperparameters for the model
param_grid = [
    {'n_estimators': [5, 10, 20, 50, 100, 200], 'max_features': [2, 4, 6, 8]},
    {'bootstrap': [False], 'n_estimators': [5, 10, 20, 50, 100, 200], 'max_features': [2, 5, 8]},
  ]

rand_forest = RandomForestRegressor(random_state=0)
grid_search = GridSearchCV(rand_forest, param_grid, cv=5,
                           scoring='neg_mean_squared_error',
                           return_train_score=True)
grid_search.fit(X_prep, y_train)
```

```python
# Getting best parameters 
grid_search.best_params_
```

```python
# Checking cv score
cv_grid = grid_search.cv_results_
min_score = min(np.sqrt(-(cv_grid["mean_test_score"])))
print(min_score)
for score, param in zip(cv_grid["mean_test_score"], cv_grid["params"]):
    print(np.sqrt(-score), param)
```

```python
# Finding best model
grid_search.best_estimator_
```

```python
# Let's try the XGBoost Model to see if we get lower RMSE
from xgboost import XGBRegressor

xg_model = XGBRegressor(random_state=0) 
xg_grid = {
    'colsample_bytree':[0.4,0.6,0.8],
    'learning_rate':[0.1, 0.05, 0.01],
    'n_estimators':[1000, 10000]
}
                    
grid_search_xg = GridSearchCV(estimator = xg_model, param_grid = xg_grid, 
                             iid=False, cv=5, scoring='neg_mean_squared_error')
grid_search_xg.fit(X_prep, y_train)
```

```python
grid_search_xg.best_params_
```

```python
cv_grid_xg = grid_search_xg.cv_results_
min_score = min(np.sqrt(-(cv_grid_xg["mean_test_score"])))
print(min_score)
for score, param in zip(cv_grid_xg["mean_test_score"], cv_grid_xg["params"]):
    print(np.sqrt(-score), param)
```

```python
grid_search_xg.best_estimator_
```

```python
from sklearn.metrics import mean_squared_error

# XGBoost is performing better so we use this for the final model
best_model = grid_search_xg.best_estimator_
X_valid_prep = preprocessor.transform(X_valid)
predictions = best_model.predict(X_valid_prep)
final_MSE = mean_squared_error(y_valid, predictions)
final_RMSE = np.sqrt(final_MSE)
final_RMSE
```
